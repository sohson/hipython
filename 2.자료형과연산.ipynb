
### 문자열

1. 인덱싱, 슬라이싱, 멤버십
2. 문자열은 immutable "불변", 대소문자 구분
3. 문자열은 코드를 지정 >> 인코딩, 디코딩, 유니코드(한글, 이모지) 가능
4. 문자열의 +. * 가능
5. 공백처리, 형식처리 등을 해야함

````python
import re
text = "파이썬 문자열   처리"
    
# If text.split("  ") 한 출력물 : ['파이썬 문자열', ' 처리']

re.sub(r"\s+"," ", text)   # 한개이상의 공백을 / 하나의 공백으로 바꿔 / text에서
````

````python
# 사칙연산자 
+, -, *, /, //, %
# 비교연산자
>, <, ==, != 
# 대입연산자 
=, +=, -=
# 논리연산자 
and, or, not, &, |, !
# 기타 
in
````
