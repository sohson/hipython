```python
import random
```

```python
exp_p = 0.5 #확률
trial_count = 10 #시행
head_count = 0 #앞면이 나온 횟수
#동저의 앞면이 나올 확률 50%
#10번 시행 시 상대도수와 확률 비교

for cnt in range(trial_count) :
    if random.random() < exp_p :
        head_count += 1        
print(f'앞면이 나온 횟수 : {head_count}')

actual_freq = head_count/trial_count

print(f'예상확률 : {exp_p*100}%\n실제결과(상대도수): {actual_freq*100}%')
```

```python
import math
```

```python
total_balls = 7
white_balls = 4
black_balls = 3
picked_cnt = 3   # white2, black1

total_case = math.comb(total_balls, picked_cnt) # 전체 경우의수 (7c3)
white_case = math.comb(white_balls, 2) # 흰공이 2개 나올 경우의수
black_case = math.comb(black_balls, 1) # 검은공이 1개 나올 경우의수
event_case = white_case * black_case # 흰공이 2개 & 검은공이 1개 나올 경우의수
prob = event_case / total_case
print(f'확률: {prob*100:.2f}%')
```

```python
#시뮬레이션
def run_simulation(trials):
    act_freq = 0
    success_cnt = 0
    sample_space = ['W','W','W','W','B','B','B'] # 표본공간
    for i in range(trials):
        drawn_balls = random.sample(sample_space,3) #sample_space에서 랜덤하게 3개를 추출
        if drawn_balls.count('W') == 2 and drawn_balls.count('B') == 1 :
            success_cnt += 1
    act_freq = success_cnt / trials # 상대도수
    return act_freq
```

```python
trials = 10
simulation_prob = run_simulation(trials)
print(f'시뮬레이션횟수 {trials}, 결과확률 {simulation_prob*100:.2f}%')
```

# 확률의 곱셈법칙

```python
# 독립사건이 동시에 일어날 확률
a_prob = 0.5 # 주사위 짝수
b_prob = 0.5 # 동전 앞면
prob_ind = a_prob * b_prob
print(f'독립사건 확률(곱) : {prob_ind:.2f}')
```

```python
# 종속사건이 동시에 일어날 확률
total = 10
win = 2
prob_A = win/total # A가 당첨될 확률 P(A)
prob_B = (win-1) / (total-1) # A다음에 B가 당첨될 확률 P(B|A)
prob_AB = prob_A * prob_B # P(A) * P(B|A)
print(f' 종속사건 AB 의 확률 : {prob_AB*100:.2f}%')
```

# 조건부 확률

```python
sample_space = list(range(1,7))
sample_space
```

```python
# 사건A: 짝수
event_A = [
    x for x in sample_space if x%2==0
]     # 리스트 comprehension = for + if : 기존 리스트를 바탕으로 새로운 리스트를 만들 때, for문과 if문을 한 줄로 줄여 쓰는 문법
event_A
```

```python
# 사건B: 4의약수
event_B = [
    x for x in sample_space if 4%x==0
]     # 리스트 comprehension = for + if : 기존 리스트를 바탕으로 새로운 리스트를 만들 때, for문과 if문을 한 줄로 줄여 쓰는 문법
event_B
```

```python
# A,B의 교집합
eventA_and_B = set(event_A) & set(event_B) # 중복되는 값은 제거해서 집합으로 나타내기
eventA_and_B
```

```python
p_A = len(event_A)/len(sample_space)
p_A_and_B=len(eventA_and_B)/len(sample_space)
p_A, p_A_and_B
```

```python
p_B_given_A = p_A_and_B/p_A
print(f'{p_B_given_A:.2f}')
```

# 기댓값 구하기

```python
x_values = list(range(1,4))
p_x = [1/6, 3/6, 2/6]
exp_value = 0   # 기댓값
for i in range(len(x_values)) :
    exp_value += x_values[i] * p_x[i]
print(f'기댓값은 {exp_value:.4f}이다')
```
