```python
tscore = int(input('점수를 입력하세요'))
A컷 = 90
B컷 = 80
C컷 = 70
D컷 = 60 

if tscore >= A컷 : 
    print ('A')
elif tscore >= B컷 :
    print ('B')
elif tscore >= C컷 :
    print ('C')
elif tscore >= D컷 :
    print ('D')
else : print ('F')
```

```python
# 중첩if문

if tscore >= A컷 : 
    print ('A')
else:
    if tscore >= B컷 :
        print ('B')
    else : 
        if tscore >= C컷 :
            print ('C')
        else : 
            if tscore >= D컷 :
                print ('D')
            else : 
                print ('F')
```

```python
# 계산기 만들기

num1, opr, num2 = input("수식을 입력하세요. (예: 20 + 20 ) : ").split()
num1=float(num1)
num2=float(num2)

result = 0
if opr == '+':
    result = num1 + num2 
elif opr == '-' :
    result = num1 - num2 
elif opr == '*' :
    result = num1 * num2 
elif opr == '/' :
    result = num1 / num2 
else :
    print('사칙연산만을 지원합니다')
print( f'{num1} {opr} {num2}의 결과는 {result}입니다.')
```

```python
# 재고관리
inventory = { '노트북' : 5, '마우스': 123, '키보드' : 0,'모니터' : 15 }

# 0 품절, 10< 부족, <50 적정, >50 충분

if inventory['노트북'] > 50 :
    print ('충분')
elif inventory['노트북'] > 10 :
    print ('적정')
elif inventory['노트북'] > 0 :
    print ('부족')
else :
    print ('품절')
```

```python
for i in inventory :
    qty = inventory[i]
    
    if qty > 50 :
        print ('충분')
    elif qty > 10 :
        print ('적정')
    elif qty > 0 :
        print ('부족')
    else :
        print ('품절')
```

```python
for name , qty in inventory.items() :
    if qty > 50 :
         status = '충분'
    elif qty > 10 :
        status = '적정'
    elif qty > 0 :
        status = '부족'
    else :
        status = '품절'
    print (f'{name} : {qty}로 {status} 상태입니다.')
```

```python
sentence = input ('문장을 입력하세요 : ')
words = ['파이썬', '코딩', '데이터', 'AI']

found_words = None

for word in words :
    if word in sentence :
        found_word = word
        break

if found_word:
    print (f'입력한 문장에서 "{found_word}" 단어를 찾았습니다!')
else :
    print ('입력한 문장에 해당 단어가 없습니다')
```

```python
# 가위바위보
import random
p = input('가위, 바위, 보 중 하나를 입력하시오 : ')
c = random.choice(['가위', '바위', '보'])
print (f'컴퓨터의 선택 : {c}')

# '승' if ( ) else '무승부' if ( ) else '패'
result = '승' if ( (p=='가위' and c=='보') or\
                 (p=='바위' and c=='가위') or\
                 (p=='보' and c=='바위') ) else \
         '무승부' if p==c else\
         '패'
print(result)
```

# break, continue, pass

```python
numbers = [ 1,3,5,7,9,11,13,15 ] 
s_value = 9
for n in numbers :
    if n == s_value :
        print (f'찾았습니다 : {n}')
       # break  >> 가장 가까운 반복문에 영향
    else :
        print(f'현재 검사한 숫자는 : {n}')
```

```python
numbers = [ 1,3,5,7,9,11,13,15 ] 
s_value = 9
for n in numbers :
    if n == s_value :
        print (f'{n//2+1}번째에서 찾았습니다 : {n}')
        break # >> 가장 가까운 반복문에 영향
    else :
        print(f'현재 검사한 숫자는 : {n}')
```

```python
numbers = [ 1,3,5,7,9,11,13,15 ] 
s_value = 9
idx = 1 # 원래 인덱스는 0으로 시작하는 경우가 많은데 이건 n번째를 나타내는 거니까 0보다 1이 적합
for n in numbers :
    if n == s_value :
        print (f'{idx}번째에서 찾았습니다 : {n}')
        break # >> 가장 가까운 반복문에 영향
    else :
        print(f'현재 검사한 숫자는 : {n}')
    idx = idx + 1 # for문 전체 한번 수행시마다 인덱스에 1을 추가하란 뜻 
```
