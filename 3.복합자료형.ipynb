## 복사 - 얕은복사 shallow copy

```python
score = [10,20,30,40,50]

new_score=score  #일부를 슬라이싱 (score[0:1]해서 넣으면 new_score에 반영이 안됨)
new_score[0]=100
print(new_score, score)
print(new_score is score)
```

## 복사 - 깊은 복사 deep copy

```python
import copy
new_score_1=copy.copy(score)
new_score_1[1]=99
print(new_score_1, score)
print(new_score_1 is score)
```

```python
students = [{'num':'1','name':'kim','kor':80,'math':90,'eng':100,},		
            {'num':'2','name':'park','kor':90,'math':60,'eng':80},		
            {'num':'3','name':'hong','kor':70,'math':70,'eng':80},		
            {'num':'4','name':'hong1','kor':70,'math':70,'eng':80},
            {'num':'5','name':'hong2','kor':70,'math':70,'eng':80}]		
```

```python
# 학생별 총점, 평균, 순위 -> 추가 출력
for s in students :
    s['total'] = s['kor']+s['math']+s['eng'] 
    s['avg']=s['total']/3
    rank = 1
    for other in students :
        if other['total']>s['total'] :
            rank += 1
        s['rank'] = rank
    
students
```

```python
# 전체 총점, 평균 출력
totals = 0
for t in students :
    totals += t['total']
print(f'전체 총점: {totals}')

avgs = 0
for d in students :
    avgs += d['avg']
sum_avg = avgs/int(students[-1]['num'])
print(f'전체 평균: {sum_avg:.2f}')
```

```python
# 과목별 평균
kor_avgs=0
for i in students :
    kor_avgs += i['kor']
kor_avg = kor_avgs / int(students[-1]['num'])
print(f'국어 과목평균: {kor_avg}')
```
