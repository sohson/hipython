```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6)) # 1행 3열, 크기


# 요일별 팁 분포 
sns.boxplot(ax=axes[0], x='day', y='tip', data=tips_df, palette='Set3')
axes[0].set_title('Tip Distribution by Day') # 0번째(첫번째) 칸에 그려라

# 시간대별 팁 분포
sns.boxplot(ax=axes[1], x='time', y='tip', data=tips_df, palette='Pastel1')
axes[1].set_title('Tip Distribution by Time')

# 요일 및 시간대별 평균 팁 
sns.barplot(ax=axes[2], x='day', y='tip', hue='time', data=tips_df, palette='Pastel2')
axes[2].set_title('Average Tip by Day and Time')

plt.tight_layout() # 그래프 간격 자동 조절 >> 제목이나 축 레이블이 겹치는 현상 방지 
```

- 토요일, 일요일의 팁 금액범위가 평일보다 높음 >> 주말에 제공되는 요리나 서비스의 퀄리티 향상에 투자하는 것이 높은 팁을 얻기에 효율적
- 점심보다 저녁의 평균 팁이 높음 >>  
- 토요일은 팁의 분포가 넓음 >> 비일상적인 소비가 많음

```python
import numpy as np

thur_lunch_m = tips_df[(tips_df['day'] == 'Thur') & (tips_df['time'] == 'Lunch')]['tip'].mean()
thur_dinner_m = tips_df[(tips_df['day'] == 'Thur') & (tips_df['time'] == 'Dinner')]['tip'].mean()

fri_lunch_m = tips_df[(tips_df['day'] == 'Fri') & (tips_df['time'] == 'Lunch')]['tip'].mean()
fri_dinner_m = tips_df[(tips_df['day'] == 'Fri') & (tips_df['time'] == 'Dinner')]['tip'].mean()

sat_lunch_m = tips_df[(tips_df['day'] == 'Sat') & (tips_df['time'] == 'Lunch')]['tip'].mean()
sat_dinner_m = tips_df[(tips_df['day'] == 'Sat') & (tips_df['time'] == 'Dinner')]['tip'].mean()

sun_lunch_m = tips_df[(tips_df['day'] == 'Sun') & (tips_df['time'] == 'Lunch')]['tip'].mean()
sun_dinner_m = tips_df[(tips_df['day'] == 'Sun') & (tips_df['time'] == 'Dinner')]['tip'].mean()

x = np.arange(4) 
width = 0.3 

plt.bar(x - width/2, [thur_lunch_m, fri_lunch_m, sat_lunch_m, sun_lunch_m], width, label='Lunch', color='skyblue')
plt.bar(x + width/2, [thur_dinner_m, fri_dinner_m, sat_dinner_m, sun_dinner_m], width, label='Dinner', color='salmon')

plt.xticks(x, ['Thur', 'Fri', 'Sat', 'Sun'])
plt.title('Average Tip by Day & Time')
plt.legend()

plt.tight_layout()
plt.show()
```
