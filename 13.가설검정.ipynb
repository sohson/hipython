```python
import numpy as np
from scipy import stats

# 1ë‹¨ê³„: ë°ì´í„° ìˆ˜ì§‘ (ê°€ìƒ ë°ì´í„° ìƒì„±)
np.random.seed(42)  # ê²°ê³¼ ì¬í˜„ì„ ìœ„í•œ ì‹œë“œ ì„¤ì •

# ìƒí™©: í†µì¦ ìˆ˜ì¹˜ê°€ ì–¼ë§ˆë‚˜ ì¤„ì–´ë“¤ì—ˆëŠ”ê°€? (ìˆ˜ì¹˜ê°€ ë†’ì„ìˆ˜ë¡ íš¨ê³¼ ì¢‹ìŒ) 
# ì‹ ê·œ ì•½ì´ ë” ì¢‹ë‹¤ê³  ê°€ì •í•œ ë°ì´í„°
group_old = np.random.normal(loc=50, scale=10, size=30)
group_new = np.random.normal(loc=55, scale=10, size=30)

print(f"[ë°ì´í„° í™•ì¸]")
print(f"ê¸°ì¡´ ì§„í†µì œ íš¨ê³¼ í‰ê· : {group_old.mean():.2f}")
print(f"ì‹ ê·œ ì§„í†µì œ íš¨ê³¼ í‰ê· : {group_new.mean():.2f}")
print("-" * 30)

# 2ë‹¨ê³„: ê°€ì„¤ ìˆ˜ë¦½ ë° ê²€ì • í†µê³„ëŸ‰ ì‚°ì¶œ
# H0: New <= Old (íš¨ê³¼ê°€ ì—†ê±°ë‚˜ ê°™ë‹¤)
# H1: New > Old (ì‹ ê·œ ì•½ì´ ë” ìš°ìˆ˜í•˜ë‹¤)

# ë…ë¦½ í‘œë³¸ T-ê²€ì • ìˆ˜í–‰
t_stat, p_value = stats.ttest_ind(
    group_new,       # ì‹ ê·œ ì•½ (ëŒ€ë¦½ê°€ì„¤ì˜ ì£¼ì¸ê³µ)
    group_old,       # ê¸°ì¡´ ì•½
    equal_var=False, # ë“±ë¶„ì‚° ê°€ì • ì•ˆ í•¨ (Welch's t-test, ì‹¤ë¬´ì ìœ¼ë¡œ ë” ì•ˆì „í•¨)
    alternative='greater' # ì¤‘ìš”: "ë” í¬ë‹¤(ìš°ìˆ˜í•˜ë‹¤)"ë¥¼ ê²€ì¦í•˜ë¯€ë¡œ ìš°ì¸¡ ê²€ì •
)
print(f"[ê²€ì • ê²°ê³¼]")
print(f"T-statistic (ì°¨ì´ì˜ í¬ê¸°): {t_stat:.4f}")
print(f"P-value (ìš°ì—°ì¼ í™•ë¥ ): {p_value:.4f}")
print("-" * 30)
```

```python
# 3ë‹¨ê³„: ìµœì¢… ê²°ì • (ë¹„ì¦ˆë‹ˆìŠ¤ ì˜ì‚¬ê²°ì •)
alpha = 0.05  # ìœ ì˜ìˆ˜ì¤€ (ë¦¬ìŠ¤í¬ í—ˆìš© ë²”ìœ„ 5%)

print(f"[ìµœì¢… íŒë‹¨]")
if p_value < alpha:
    print(f"ê²°ê³¼: ê·€ë¬´ê°€ì„¤ ê¸°ê° (P-value {p_value:.4f} < {alpha})")
    print("í•´ì„: ì‹ ê·œ ì§„í†µì œì˜ íš¨ëŠ¥ì´ í†µê³„ì ìœ¼ë¡œ ìœ ì˜ë¯¸í•˜ê²Œ ìš°ìˆ˜í•©ë‹ˆë‹¤.")
    print("ğŸ‘‰ Action: ì‹ ê·œ ì•½ ì¶œì‹œ ìŠ¹ì¸ / ë§ˆì¼€íŒ… í¬ì¸íŠ¸ë¡œ í™œìš©")
else:
    print(f"ê²°ê³¼: ê·€ë¬´ê°€ì„¤ ì±„íƒ (P-value {p_value:.4f} >= {alpha})")
    print("í•´ì„: í‰ê· ì€ ì¡°ê¸ˆ ë†’ì•„ ë³´ì¼ì§€ ëª°ë¼ë„, í†µê³„ì ìœ¼ë¡œ ìš°ìˆ˜í•˜ë‹¤ê³  ë³¼ ìˆ˜ ì—†ìŠµë‹ˆë‹¤ (ìš°ì—°ì¼ ê°€ëŠ¥ì„± ë†’ìŒ).")
    print("ğŸ‘‰ Action: ì¶œì‹œ ë³´ë¥˜ / ì¶”ê°€ ì„ìƒ ì‹¤í—˜ í•„ìš”")
```

```python
# ë‹¨ì¼í‘œë³¸ t-ê²€ì • : í•˜ë£¨ì»¤í”¼ì„­ì·¨ëŸ‰ì´ í‰ê·  2.5ì”ì´ë¼ê³  ì•Œë ¤ì ¸ ìˆê³  ì‹¤ì œì¡°ì‚¬ì—ì„œ 2.8ì”ì´ ë‚˜ì™”ë‹¤. 
# H0: ì°¨ì´ê°€ í†µê³„ì ìœ¼ë¡œ ìœ ì˜í•˜ì§€ì•Šë‹¤ / H1: ì°¨ì´ê°€ í†µê³„ì ìœ¼ë¡œ ìœ ì˜ë¯¸í•˜ë‹¤.
coffee_data = [2.9, 2.4, 2.7, 3.0, 2.6, 2.8, 2.9, 3.1, 2.7, 2.6,
               2.8, 2.5, 3.2, 3.0, 2.4, 2.6, 2.7, 3.1, 2.8, 2.7,
               2.5, 2.9, 2.6, 2.8, 2.9, 2.5, 2.7, 2.8, 3.0, 2.9]

t_stat, p_value = stats.ttest_1samp(coffee_data, popmean=2.5) # 1sampì´ë‹ˆê¹Œ ë‹¨ì¼í‘œë³¸ ê²€ì •
print(f"tê°’ = {t_stat:.3f}, pê°’ = {p_value:.3f}")  # pê°’ì´ 0.05ë³´ë‹¤ ì‘ìœ¼ë‹ˆê¹Œ ê·€ë¬´ ê¸°ê° > ì°¨ì´ê°€ ìœ ì˜ë¯¸í•˜ë‹¤.
```

```python
# ë…ë¦½í‘œë³¸ T ê²€ì • : ë‘ ê°œì˜ ëª¨ì§‘ë‹¨ì˜ í‰ê·  ë¹„êµ 
# H0: ë‘ í‘œë³¸ í‰ê· ì ìˆ˜ì— ìœ ì˜ë¯¸í•œ ì°¨ì´ê°€ ì—†ë‹¤ / H1: ì°¨ì´ê°€ ìˆë‹¤
group_A = [78, 85, 82, 88, 76, 81, 79, 84, 82, 80]
group_B = [72, 75, 78, 74, 71, 77, 70, 76, 75, 73]

t_stat, p_value = stats.ttest_ind(group_A, group_B, equal_var=False) 
# independentì˜ ì•½ìë¡œ ë…ë¦½ì ì¸ ë‘í‘œë³¸ì„ ê²€ì •, equal_var=falseëŠ” ë‘ í‘œë³¸ ë¶„ì‚°ì„ ë‹¤ë¥´ê²Œ ì§€ì • 
print(f"tê°’ = {t_stat:.3f}, pê°’ = {p_value:.3f}") # pê°’ì´ 0.05ë³´ë‹¤ ì‘ìœ¼ë‹ˆê¹Œ ê·€ë¬´ ê¸°ê° > ì°¨ì´ê°€ ìœ ì˜ë¯¸í•˜ë‹¤.
```

```python
# ëŒ€ì‘í‘œë³¸(ë™ì¼ì§‘ë‹¨ì˜ ì „í›„ì°¨ì´ ë¹„êµ) t-ê²€ì • ì˜ˆì œ: ì•½ë¬¼ íˆ¬ì—¬ ì „í›„ í˜ˆì•• ë¹„êµ
# H0: ë‘ í‘œë³¸ í‰ê· ì ìˆ˜ì— ìœ ì˜ë¯¸í•œ ì°¨ì´ê°€ ì—†ë‹¤ / H1: ì°¨ì´ê°€ ìˆë‹¤
before = [120, 122, 143, 130, 135, 118, 127, 140, 132, 125]
after = [115, 117, 138, 125, 129, 112, 120, 135, 126, 119]

t_stat, p_value = stats.ttest_rel(before, after) # relì´ë‹ˆê¹Œ ëŒ€ì‘í‘œë³¸ ê²€ì •
print(f"tê°’ = {t_stat:.3f}, pê°’ = {p_value:.3f}") # pê°’ì´ 0.05ë³´ë‹¤ ì‘ìœ¼ë‹ˆê¹Œ ê·€ë¬´ ê¸°ê° > ì°¨ì´ê°€ ìœ ì˜ë¯¸í•˜ë‹¤.
```
