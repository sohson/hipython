```python
def 계산기(*a) :   # '*'를 붙이면 갯수 상관없이 넣을 수 있음
    print(a)
    return sum(a)
```

```python
def 계산기(a,b,*c) :
    sum_all = a+b+sum(c)
    return sum_all
```

```python
계산기(1,2,10)
```

# CRUD : 간식박스 프로그램

1. 함수 도출 : CRUD 기본
2. 함수 구현
3. 문제점 도출 -> 함수 리팩토링
4. 전체흐름을 구현하는 함수 구현

```python
# 간식박스 생성
def make_box():
    snackbox = ['두쫀쿠']
    return snackbox
    
# 간식박스 추가
def add_box(sbox):
    make_msg = input('먹고싶은 간식을 입력하세요. (단, 여러개인 경우 쉼표로 구분하여 입력)').split(",")
    sbox.extend(make_msg)
    #print(sbox)
    return sbox

# 전체 보기
def display_box(sbox) :
    print(sbox)
    
# 선택지 출력
def display_menu():
    print("1. 간식추가")
    print("2. 간식박스 확인")
    print("3. 간식제거")
    print('4. 주문확정')

# 주문 확정
def order_confirm():
    make_msg = input('주문을 확정하시겠습니까? (yes/no)') #다른함수 안에있는 동일변수 make_msg는 해당함수에서만 적용되는 
                                                         #지역변수이므로 겹쳐도 상관없음 
    if make_msg == 'yes' :
        return True
    else : 
        return False

# 삭제
```

```python
# 서비스의 흐름
snack_box = make_box()
print('간식박스 생성완료', snack_box)
while True :
    display_menu()
    menu_no = input('번호를 입력하세요')
    if menu_no == '1' :
        add_box(snack_box)
    elif menu_no =='2' :
        display_box(snack_box)
    elif menu_no =='4' :
        display_box(snack_box)       # 한번 보여주고
        result =  order_confirm()    # 끝
        if result :
            break
        else :
            continue

print('주문 감사드립니다.')
```
