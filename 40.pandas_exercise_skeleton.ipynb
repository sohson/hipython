# Q10*
charges의 표시형식을 소수점 아래 2자리로 바꾸기

```python
# 각 요소에 f-string 형식을 적용
#insurance_df['charges'].map(lambda x: f'{float(x):.2f}')

# 데이터를 실제로 반올림하여 변경
# insurance_df['charges'] = insurance_df['charges'].round(2)
# insurance_df['charges'] = round(insurance_df['charges'], 2)

# apply 함수 : input과 return 정의 >> 커스텀함수를 df에 적용
def round_2(data) :
  return round(data, 2)
insurance_df['charges'] = insurance_df['charges'].apply(round_2)

insurance_df
```
