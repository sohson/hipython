```python
import pymysql
```

```python
# MYSQL 서버에 연결
con = pymysql.connect(
  host = 'localhost',
  user = 'root',
  password ='1234',
  db='wntrade',
  charset='utf8' 
) #한글처리 (charset ='utf8')
# 커서 생성
cursor = con.cursor()
```

```python
cursor
```

```python
cursor.execute("show tables")
```

```python
for row in cursor.fetchall():
  print(row)
```

```python
# SQL 쿼리 실행
sql = """
SELECT 사원번호, 이름, 부서명
FROM 사원 LEFT JOIN 부서
ON 사원.부서번호 = 부서. 부서번호
"""
```

```python
cursor.execute(sql)
```

```python
for row in cursor.fetchall():
  print(row)
```

```python
cursor.close()
con.close()
```
