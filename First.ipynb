```python
current_value = 200000
annual_depreciation_rate = 0.1

def calculate_depreciation(asset_age):
    global current_value 
    for i in range(asset_age):
        current_value = current_value * (1- annual_depreciation_rate)
    return current_value

calculate_depreciation(3)
print(current_value)

```

```python
활동_목록 = [{'a':'제품 판매', 'c':'수익','d':20250228,'w':'토'},
         {'a':'복사기', 'c':'비용','d':20250305,'w':'일'},{'a':'복사기', 'c':'아님','d':20250310,'w':'일'},
         {'a':'기부금', 'c':'비용','d':20250302,'w':'토'},{'a':'온라인', 'c':'수익','d':20250304,'w':'수'}]

from datetime import datetime
w_idx={'월':0, '화':1, '수':2, '목':3, '금':4,'토':5,'일':6}
회계상_거래=['수익','비용']

회계_항목=list(filter(lambda 항목: 항목['c'] in 회계상_거래 
                  and datetime.strptime(str(항목['d']), '%Y%m%d')>datetime(2025,3,1) 
                  and w_idx[항목['w']] in [5,6], 활동_목록))

결과 = {}
for 항목 in 회계_항목:
    분류 = 항목['c']
    결과[분류] = 결과.get(분류,0) +1
print(결과)
```
