#  이전대화를 포함해서 답변하기

- system
- user
- assistant : 그전에 기억하고 싶은 대화를 저장 >> RAG, 검수 by 사람, 중가산출물

````python
response = client.responses.create(
    model="gpt-4o-mini",
    input=[
        msg("user", "2002년 월드컵에서 가장 화제가 되었던 나라는 어디야?"),
        msg("user", "그나라가 화제가 되었던 이유는? ")
    ],
    temperature=0
)
# system에 대한 지정이 없더라도  API는 정상작동함
response.output_text
````

````python
response = client.responses.create(
    model="gpt-4o-mini",
    input=[
        msg("user","2002년 월드컵에서 가장 화제가 되었던 나라는 어디야?"),
        {
            "role": "assistant", #헬퍼는 그대로 두고, assistant만 명시적으로 처리
            "content": [
                {
                    "type": "output_text",
                    "text": "바로 예상을 뚫고 우승한 브라질입니다."
                }
            ]
        },
        msg("user","그 나라가 화제가 되었던 이유를 자세하게 설명해줘")
    ]
)
# 구조: User 질문 1 -> Assistant: 강제 주입(브라질) -> User 질문 2
response.output_text
````

````python
def msg_assist(role, text) :
    msg_src = {"role": role, "content": [{"type":"output_text", "text":text}] }
    return msg_src
````

````python
response = client.responses.create(
    model="gpt-4o-mini",
    input=[
        msg("user","2002년 월드컵에서 가장 화제가 되었던 나라는 어디야?"),
        msg_assist("assistant","바로 예상을 뚫고 우승한 브라질입니다." ),
        msg("user","그 나라가 화제가 되었던 이유를 자세하게 설명해줘")
    ]
)
response.output_text
````

# 파라미터의 제어
- temperature : 창의성 0~1 (최대)
- max_output_tokens : 최대 응답 길이
- top_p : 후보선정 기준 설정 즉, 높을수록 선택지의 양이 많다고 생각 >> 높을수록 다양하고 창의적 답 cf. 아예 커트라인을 자른다는게 temp랑 차이
- frequency_penalty : 반복 억제

````python
top_p = [0.6, 0.8, 1.0]  # top_p 조건이랑 temperature 조건은 같이 안쓰는게 좋음 (조건이 꼬이지않게)
for p in top_p:
    response = client.responses.create(
        model="gpt-4o-mini",
        input="마케팅 문구 1개 만들어줘",
        #temperature=0.5,
        top_p=p
    )
    print(response.output_text)
````
