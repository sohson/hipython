```python
# 양수음수분류기: 다양한 개수의 정수를 입력받아, 각 수가 양수인지 음수인지 구분하여 각각의 개수 및 합계를 출력
# 입력 받은 값 중 0은 무시
# quit 를 입력하면 프로그램을 중지

positive = 0
negative = 0
sum = 0
num_list = []
p_list = []
n_list = []

while True :
    num = input("정수를 입력해주세요(종료하려면 quit 입력) : ")
    if num == 'quit':
        break

    n = int(num)
    sum += n
    if n > 0 :
        positive += 1
        num_list.append(n)
        p_list.append(n)
    elif n < 0 :
        negative += 1
        num_list.append(n)
        n_list.append(n)
    else  :
        print('입력한 숫자가 없습니다.')
   
print (f'카운팅 종료. \n{num_list} 중 \n양수: {positive}개 {p_list}, \n음수: {negative}개 {n_list}, \n합계: {sum}')//
```

```python
#초대장 생성 프로그램

vip_list = ['조인성', '김우빈', '정우성', '유해진']


while True:
    vip = input("vip 이름을 입력하세요 : ")

    if vip == '조인성' or vip == '김우빈' or vip == '정우성' or vip == '유해진' :
        idx = vip_list.index(vip)
        print (f''' 
        {vip}님 안녕하세요? \n
        화창한 2023년 가을, 저희 컴퓨터 동아리 축제에 {vip}님을 초대합니다. \n
        쿠폰 No. {idx+1}
        {vip}님 축제에서 만나길 고대하겠습니다.
        ''')
    else :
        print(f'{vip}님은 초대받지 못하셨습니다. 죄송합니다.')
        break
    
```

```python
#초대장 생성 프로그램 (enumerate 활용)

vip_list = ['조인성', '김우빈', '정우성', '유해진']


while True:
    vip = input("vip 이름을 입력하세요 : ")
    if vip in vip_list :
        for v, i in enumerate(vip_list) :
            if vip == i :
                print (f''' 
            {i}님 안녕하세요? \n
            화창한 2023년 가을, 저희 컴퓨터 동아리 축제에 {i}님을 초대합니다. \n
            쿠폰 No. {v+1}
            {i}님 축제에서 만나길 고대하겠습니다.
              ''')
                break 
    else :
        print(f'{vip}님은 초대받지 못하셨습니다. 죄송합니다.')
        break
    
```
