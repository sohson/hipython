```python
import pandas as pd

# 직원 명단 DataFrame 생성
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 45, 35, 32, 28],
    'Salary': [5000, 7000, 6500, 5500, 6000]
}
sal_df = pd.DataFrame(data)
sal_df.info()
```

```python
def age_category(age):
  if age < 30 :
    return '20대'
  elif age < 40 :
    return '30대'
  else :
    return '60대 이상'
```

```python
sal_df['Age_category'] = sal_df['Age'].apply(age_category)
sal_df
```

```python
# 보너스 칼럼 20%
sal_df['Bonus'] = sal_df['Salary'].apply(lambda x: x * 0.2)
sal_df
```

```python
df_new = df.drop(columns='등록일자')
df_new
```

```python
df_new.info()
```

```python
# 등록일자를 int에서 날짜형으로
# df_new['사용일자']

def changeDate(data):
  year=str(data)[0:4]
  month=str(data)[4:6]
  day=str(data)[6:8]
  return year+'-'+month+'-'+day

df_new['사용일자']=df_new['사용일자'].apply(changeDate)
df_new['사용일자']=pd.to_datetime(df_new['사용일자'])
df_new.head()

df_final = df_new.drop(columns='사용일자1')
df_final
```

```python
# 2호선 추출 후 역별 승하차총승객수차이, 승하차총승객수
df_line2 = df_final[df_final['노선명']=='2호선']

df_line2_역별 = df_line2.groupby('역명')[['승차총승객수', '하차총승객수']].sum().reset_index()
df_line2_역별['승하차총승객수차이'] = df_line2_역별['승차총승객수'] - df_line2_역별['하차총승객수']

df_line2_역별['승하차총승객수'] = df_line2_역별['승차총승객수'] + df_line2_역별['하차총승객수']

top10 = df_line2_역별.sort_values('승하차총승객수',ascending=False).head(10)
top10
```

```python
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

plt.subplot(1,2,1)
plt.bar(top10['역명'], top10['승하차총승객수'],color='skyblue')
plt.xticks(rotation=45)
plt.title('승하차총승객수')

plt.subplot(1,2,2)
plt.bar(top10['역명'], top10['승하차총승객수차이'], color='pink')
plt.xticks(rotation=45)
plt.title('승하차총승객수차이(승차-하차)')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

plt.subplot(1,2,1)
plt.bar(top10['역명'], top10['승하차총승객수'],color='skyblue')
plt.xticks(rotation=45)
plt.title('승하차총승객수')

plt.subplot(1,2,2)
plt.plot(top10['역명'], top10['승하차총승객수'], marker='o')
plt.xticks(rotation=45, ha='right')
plt.show()

```

```python
plt.figure(figsize=(7,5))
plt.scatter(
    top10['승하차총승객수'],
    top10['승하차총승객수차이']
)
plt.axhline(0)  # 기준선
plt.title('2호선 Top 10 역: 규모 vs 승하차 차이')
plt.xlabel('승하차총승객수')
plt.ylabel('승하차총승객수차이')


# 역명 라벨
for i in range(len(top10)):
    plt.text(
        top10['승하차총승객수'].iloc[i],
        top10['승하차총승객수차이'].iloc[i],
        top10['역명'].iloc[i],
        fontsize=9
    )
plt.show()
```
