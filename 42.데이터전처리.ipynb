```python
import pandas as pd

# 직원 명단 DataFrame 생성
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 45, 35, 32, 28],
    'Salary': [5000, 7000, 6500, 5500, 6000]
}
sal_df = pd.DataFrame(data)
sal_df.info()
```

```python
def age_category(age):
  if age < 30 :
    return '20대'
  elif age < 40 :
    return '30대'
  else :
    return '60대 이상'
```

```python
sal_df['Age_category'] = sal_df['Age'].apply(age_category)
sal_df
```

```python
# 보너스 칼럼 20%
sal_df['Bonus'] = sal_df['Salary'].apply(lambda x: x * 0.2)
sal_df
```

## concat
시리즈를 행으로 만들어서 concat
- 원본과 동일한 컬럼을 갖는 empty dataframe 생성
- '칼럼이름 = 인덱스이름'인 시리즈 생성
- empty dataframe에 행으로 시리즈를 추가
- concat(원본, empty dataframe) 

```python
df_1=pd.read_csv('./data/concat_1.csv')
df_2=pd.read_csv('./data/concat_2.csv') 
df_3=pd.read_csv('./data/concat_3.csv')
```

```python
total_df = pd.concat([df_1,df_2,df_3], ignore_index=True)
total_df
```

새로운 행 추가

```python
add_ser = pd.Series(['a100', 'b100', 'c100','d100'])
add_ser
```

```python
total_df_add = pd.concat([total_df,add_ser])
# total_df_add
```

```python
col_list = total_df.columns.tolist()
```

```python
# dataframe으로 만든다.
empty_df = pd.DataFrame(columns = col_list)
# empty_df.loc[0] = add_ser
empty_df
```

```python
add_ser.index= empty_df.columns
add_ser
```

```python
empty_df.loc[0] = add_ser
empty_df
```

```python
new_total_df= pd.concat([total_df,empty_df], ignore_index=True)  
new_total_df
```

## merge
- 키를 기준으로 합친다.
- sql 에서 join

```python
# 매출 데이터 
sales_data = pd.DataFrame({ 'CustomerID': [101, 102, 103, 104], 'Date': ['2023-01-01', '2023-01-03', '2023-01-05', '2023-01-07'], 'Amount': [250, 300, 150, 400] })
# 고객 데이터
customer_data = pd.DataFrame({ 'CustomerID': [101, 102, 103, 105], 'Name': ['Alice', 'Bob', 'Charlie', 'Eva'], 'Region': ['North', 'South', 'East', 'West'] })
```

```python
sales_data.info()
```

```python
customer_data.info()
```

```python
pd.merge(customer_data, sales_data, on='CustomerID', how='left')
 # (left, right, 기준키, 조인방식 {default=inner})
```

```python
df=pd.DataFrame(data)
df
```

```python
df.info()
```

```python
df.isnull()
```

```python
df.isnull().sum() # null 값의 갯수
```

```python
df.dropna()
```

```python
# null 값 >> 0으로 대체
df.fillna(0)
```

```python
df['age'].fillna(f'{df["age"].mean():.2f}')
```
