```python
# 문자열 순회
password = 'Abc21!'

# 비밀번호 조건 만족여부 확인
isUpper = False
isDigit = False
isSpecial = False
isLower = False
special_char = '!@#%?$'

for char in password :                                   # char은 password에서 A, b, 2, ! 와 같이 한글자씩 떼서 넣어두는 거임
    if char.isupper() : # 대분자 확인 
        isUpper = True
    elif char.isdigit() : # 숫자 확인
        isDigit = True
    elif char in special_char :   # 특수기호 확인          # char 안에 특수기호가 있니?
        isSpecial = True
    elif char.islower() :        # 소문자 확인
        isLower = True
    check_result = isUpper + isDigit + isSpecial + isLower
    
print(f"비밀번호 강도 조건 만족여부: {'만족' if check_result == 4 else '불만족'}")
```

```python
# 년월일을 입력받아 요일 출력하기
year = int(input('몇 년: '))
month = int(input('몇 월: '))
day = int(input('몇 일: '))
week_day = ['월', '화', '수', '목', '금', '토', '일']

import datetime
wd = datetime.date(year,month,day).weekday()
print(f'{week_day[wd]}요일 입니다')
```

```python
# 총일수 % 7 -> wd
# 기준일 : 2025-10-5
# 연도 : 1~2024 년도 수 * 365/6
# 월 : 1~9월 수 * 31/30/29/28
# 일 : 일
```

```python
year = int(input('연도를 입력하시오: '))
month = int(input('월을 입력하시오: '))
day = int(input('일을 입력하시오: '))
```

```python
year, month, day
```

```python
total_days=0
# 연도의 총 일수 계산
for i in range(1, year) : #2024년까지 셈
    if i % 400 == 0 : #윤년
        total_days += 366  #365를 더하는 과정을 range(1,year)번 반복해서 i에 저장함
    elif i % 100 == 0 : #평년
        total_days += 365
    elif i % 4 == 0 : #윤년
        total_days += 366
    else : total_days += 365
print(f'1st : {total_days}, {year}')

# 기준일까지의 2025년 월의 총 일수 계산
month_days = [0,31,28,31,30,31,30,31,31,30,31,30,31]
for m in range(1,month) :
    total_days += month_days[m]
print(f'2nd : {total_days}')

# 기준월의 일수 추가
total_days += day
if month >= 3 :
    #윤년체크
    if year % 400 == 0 : #윤년
        total_days += 1  
    elif year % 100 == 0 : #평년
        total_days += 0
    elif year % 4 == 0 : #윤년
        total_days += 1 
    else : total_days += 0
print(f'3rd : {total_days}')

# 연도 + 월 + 일 % 7 = 요일에 해당하는수 
remain = total_days % 7
result = None 
if remain == 0:
    result ='일'
elif remain == 1:
    result ='월'
elif remain == 2:
    result ='화'
elif remain == 3:
    result ='수'
elif remain == 4:
    result ='목'
elif remain == 5:
    result ='금'
elif remain == 6:
    result ='토'

print(f'{year}년 {month}월 {day}일은 {result}요일 입니다.')
```

```python
year = int(input('몇 년: '))
month = int(input('몇 월: '))
day = int(input('몇 일: '))
total_days = 0

# 1. 서기 1년부터 작년(year-1)까지의 총 일수 합산
for y in range(1, year):
    # 윤년 조건: 4로 나누어 떨어지면서 100으로 안 나누어지는 연도에 예외적으로 400으로 나누어 떨어지는 연도까지
    if (y % 4 == 0 and y % 100 != 0) or (y % 400 == 0):
        total_days += 366
    else:
        total_days += 365

# 2. 올해 1월부터 지난달(month-1)까지의 일수 합산
for m in range(1, month):
    if m in [1, 3, 5, 7, 8, 10, 12]:
        total_days += 31
    elif m in [4, 6, 9, 11]:
        total_days += 30
    elif m == 2:
        # 올해가 윤년인지 확인하여 2월 일수 결정
        if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
            total_days += 29
        else:
            total_days += 28

# 3. 해당 기준월의 일수 추가
total_days += day

# 4. 요일 리스트 및 결과 출력
week_day = [ '일', '월', '화', '수', '목', '금', '토' ]
wd_index = total_days % 7
2
print(f'총 {total_days}일')
print(f'{week_day[wd_index]}요일 입니다')
```
